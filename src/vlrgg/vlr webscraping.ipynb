```python
events_url = 'https://www.vlr.gg/events/'
event_root = 'https://www.vlr.gg/event/'
```

```python
import bs4

def _event_url(event, events_resp):
    url_string = None
    if re.match('(\d+)', event):
        url_string = event_root + event
    else:
        if events_resp:
            event_text = BeautifulSoup(events_resp.text, 'lxml')
            event_items = (event_text.find_all('a', 'event-item'))
            for event_item in event_items:
                title_tag = event_item.find('div', class_='event-item-title')
                item_title = title_tag.contents[0].strip()
                if event == item_title:
                    url_string = event_root +
                    break

    return url_string
```

```python
event_id = '926'
main_resp = requests.get(vlr_events_url)
url = _event_url(event_id, main_resp)
matches_url = matches_url = (url.replace('event', 'event/matches')
                       + '/&group=completed')
matches_resp = requests.get(matches_url)
urls = _event_match_urls(matches_resp)
urls
```

```python
def _event_match_urls(matches_resp):
    urls = []
    if matches_resp:
        matches_soup = BeautifulSoup(matches_resp.text, 'lxml')
        match_items = matches_soup.find_all('a', 'match-item')
        urls = [match_item['href'] for match_item in match_items]

    return urls
```
